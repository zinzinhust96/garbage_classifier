```python
from PIL import Image
from torchvision import transforms
from IPython.display import display
```

```python
# overall pipeline
transform = transforms.Compose([
    transforms.RandomResizedCrop(394, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    # transforms.RandomRotation(30),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
# read an image using PIL
img = Image.open('/home/namdng/garbage_classfier/data_split/test/others/battery_27_data2.jpg')
img.size
```

```python
# show the original image
display(img)
```

```python
# Resize
resized_transform = transforms.Resize(394)
resized_img = resized_transform(img)
display(resized_img)
```

```python
# RandomResizedCrop
random_resized_crop_transform = transforms.RandomResizedCrop(394, scale=(0.8, 1.0))
random_resized_crop_img = random_resized_crop_transform(img)
display(random_resized_crop_img)
```
